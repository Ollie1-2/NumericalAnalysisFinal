```python
def f(x,t): return x*sin(x)
time = np.linspace(0,15,100)
sol = odeint(f,0.1,time)
sol[:8]
```

```python
plt.plot(time,sol)
plt.plot([0,15],[3,3])
```

```python
interpolation = interp1d(time,[y[0] for y in sol], kind=3)
def f(x): return interpolation(x)-3

brentq(f,0,12)
```

```python
def midpoint_sum(f,a,b,n):
    xs = np.linspace(a,b,n+1)
    ys = f(xs)
    return sum([(ys[i]+ys[i-1]) for i in range(n)])*(b-a)/(2*n)
midpoint_sum(ff, 0, 1, 500)
```
